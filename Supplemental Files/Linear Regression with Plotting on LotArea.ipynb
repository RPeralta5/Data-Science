```python
# Splitting the dataset into testing and training:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, 
                                                    random_state=2)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)

print(data_set.dtypes)
```

```python
# In the following line, "my_linreg" is instantiated as an "object" of LinearRegression "class". 
my_linreg = LinearRegression()

# fitting the model to the training data:
my_linreg.fit(X_train, y_train)
```

```python
# printing Theta0 using attribute "intercept_":
print(my_linreg.intercept_)

# printing [Theta1, Theta2, Theta3] using attribute "coef_":
print(my_linreg.coef_)
```

```python
# Make predictions on the testing set
y_prediction = my_linreg.predict(X_test)

print(y_prediction)
```

```python
# RMSE

from sklearn import metrics
import numpy as np

# Calculating "Mean Square Error" (MSE):
mse = metrics.mean_squared_error(y_test, y_prediction)

# Using numpy sqrt function to take the square root and calculate "Root Mean Square Error" (RMSE)
rmse = np.sqrt(mse)

print(rmse)

#print(X.size)
#print(y.size)
```

```python
# Plotting using LotArea as X
from numpy import *

# Take out outliers
data_set = data_set[data_set['LotArea'] < 50000]  


from scipy.interpolate import *

X = data_set['LotArea']
y = data_set['SalePrice']
print(X.ndim)
p1 = polyfit(X,y,1)
print(p1)


import operator
import matplotlib.pyplot as plt
plt.scatter(X, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(X,y), key=sort_axis)
X, y = zip(*sorted_zip)
plt.plot(X, y, color='r')
plt.xlabel("Lot Area")
plt.ylabel("SalesPrice")
plt.show()

plt.title("Linear Regression 1st Order")
p1 = polyfit(X,y,1)
plt.scatter(X, y, s=10)
plt.plot(X,polyval(p1,X))
plt.xlabel("Lot Area")
plt.ylabel("SalesPrice")
plt.show()

plt.title("Polynomial Regression 2nd Order")
p2 = polyfit(X,y,2)
plt.scatter(X, y, s=10)
plt.plot(X,polyval(p2,X))
plt.xlabel("Lot Area")
plt.ylabel("SalesPrice")
plt.show()

plt.title("Polynomial Regression 3rd Order")
p3 = polyfit(X,y,3)
plt.scatter(X, y, s=10)
plt.plot(X,polyval(p3,X))
plt.xlabel("Lot Area")
plt.ylabel("SalesPrice")
plt.show()
```
